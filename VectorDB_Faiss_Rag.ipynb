```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
import openai

sentences = [
    "I love cats",
    "Dogs are best",
    "Artificial Intellignce is the future",
    "I enjoy machine learning",
    "I have a pet dog",
    "Cats are cute animals"
]
```

```python
#Load pre-trained model embeeding data
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(sentences)  #Shape:(6,384)
embeddings = np.array(embeddings).astype("float32")
```

```python

embeddings.shape
```

```python
dimension = embeddings.shape[1] # 1 will have 384
index = faiss.IndexFlatL2(dimension)     #This is actually the vectorDB functioning
index.add(embeddings) #finding the ecuclieden distance
```

```python
query = "I adore kittens"
query_vector = model.encode([query]).astype("float32")

top_k=3
distances,indices = index.search(query_vector,top_k)

print("Query:",query)
print("\nTop matches:")
for idx in indices[0]:
    print("-",sentences[idx])
```
